```python
#Write a Python program that checks whether a given number is prime or not. A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.
def is_prime(num):
    if num == 2:
        return True
    if num <= 1:  # Separate check for numbers less than or equal to 1
        return False
    if num % 2 == 0:  # Separate check for even numbers greater than 2
        return False
    # to make the range between 3 to half of the number 
    for i in range(3,int(num ** 0.5) + 1,2):
        if num % i == 0:
            return False
    return True
num = int(input("Enter the number"))
if is_prime(num):
    print("it is a prime number")
else:
    print("It is not a prime number")
```

```python
#Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers.
#The program should then check if the user's answer is correct and display an appropriate message.
import random
# Define the function
def main():
    #Generate Two random numbers
    num1 = random.randint(1,10)
    num2 = random.randint(1,10)
    #Product of the two random numbers
    prod = num1 ** num2
    #Ask user for the random number
    num = int(input("enter the product"))
    #Check if both are equal or not
    if prod == num:
        print("Excellent ")
    else:
        print("not equal")
# running the main function
if __name__ == '__main__' :
    main()

        
```

```python
#write a program to count the number of words in a given text.
def count_words(text):
    text.lower()
    words = text.split()
    word_count={}
    #counting the words
    for word in words:
        word.strip(",.!@#$%^&*(){}[]")
        if word:
            word_count[word] = word_count.get(word,0) + 1
    #Returning the word_count dictionary
    return word_count
#Sample text
Sample_text = "This is a sample  text . This is just for fun"
result = count_words(Sample_text)

#Printing the Result
for word,count in result.items():
    print(f"'{word}':{count}")
```

```python
#Write a Python function called is_palindrome that takes a string as input and returns True if the string is a palindrome, 
#and False otherwise. A palindrome is a word, phrase, number, 
#or other sequence of characters that reads the same forward and backward, ignoring spaces, punctuation, and capitalization.
import string
def is_pallindrom(s):
    cleaned_string = ''.join(char.lower() for char in s if char.isalnum())
    return cleaned_string == cleaned_string[::-1]
#example usage
test_string = 'racecar'
if is_pallindrom(test_string):
    print(f'"{test_string}" is a pallindrom')
else:
    print(f'"{test_string}" is a pallindrom')      
```
