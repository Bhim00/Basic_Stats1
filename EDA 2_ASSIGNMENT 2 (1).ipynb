```python
columns = ['age','fnlwgt','education_num','capital_gain','capital_loss','hours_per_week']
for column in columns:
    if column in df.columns:
        df[column] = df[column].fillna(df[column].median())
```

```python
df[columns].isnull().sum()
```

```python
for column in columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR
    df[column] = np.where((df[column] < lower_bound)|(df[column] > upper_bound),df[column].median(),df[column])
```

```python
df.describe()
```

```python
std = StandardScaler()
df[columns] = std.fit_transform(df[columns])
```

```python
scaler = MinMaxScaler()
df[columns] = scaler.fit_transform(df[columns])
```

```python
categorical_columns = [
    col for col in df.select_dtypes(include='object').columns if df[col].nunique() < 5
]
```

```python
print("Categorical columns with less than 5 unique values:")
print(categorical_columns)

# Appling one-hot encoding only to the identified categorical columns
df_encoded = pd.get_dummies(df, columns=categorical_columns, drop_first=False)
```

```python
df_encoded
```

```python
#LabelEncoding
categorical_columns = [
    col for col in df_encoded.select_dtypes(include='object').columns if df_encoded[col].nunique() >= 5
]
label_encoder = LabelEncoder()

for col in categorical_columns:
    df_encoded[col] = label_encoder.fit_transform(df_encoded[col])
```

```python
df_encoded
```

```python
#Creating new Features
# Income_Per_Hour - The feature adjusts income based on hours worked, providing a normalized measure of income. This helps compare individuals with different work hours on a consistent basis.
df_encoded['Income_Per_Hour'] = (df_encoded['capital_gain'] - df_encoded['capital_loss']) / df_encoded['hours_per_week']
```

```python
df_encoded
```

```python
# This feature can aid in segmenting populations based on work-life priorities, which is useful in sociological or economic studies, or for developing personalized recommendations or interventions.
df_encoded['Work_Life_Balance'] = df_encoded['hours_per_week'] / df_encoded['age']
```

```python
df_encoded
```

```python
#Applying IsolationForest
iso_forest = IsolationForest(contamination=0.05) 
columns = ['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']
df_numeric = df_encoded[columns]
```

```python
outliers = iso_forest.fit_predict(df_numeric)
```

```python
df['outlier'] = outliers
```

```python
df_no_outliers = df[df['outlier'] == 1].drop('outlier', axis=1)
```

```python
print(f"Before removing outliers: {len(df)}")
print(f"After removing outliers: {len(df_no_outliers)}")
```
