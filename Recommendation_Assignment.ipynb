```python
#Importing Libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.model_selection import train_test_split
```

```python
#Loading Dataset
data = pd.read_csv("C:\\Users\\bhimr\\OneDrive\\Documents\\Recommendation System\\anime.csv")
new_data = data.copy()
```

```python
data1 = data.drop(['anime_id'],axis = 1)
```

```python
unique_1 = data['genre'].unique()
print(unique_1)
unique_2 = data['type'].unique()
print(unique_2)
```

```python
# Step 1: Ensure 'genre_list' exists and contains lists
#df['genre'] = df['genre'].fillna('')  # Handle NaN values
data1['genre_list'] = data1['genre'].apply(lambda x: str(x).split(', '))  # Convert to lists

# Step 2: Apply MultiLabelBinarizer
mlb = MultiLabelBinarizer()
genre_encoded = mlb.fit_transform(data1['genre_list'])

# Step 3: Convert the encoded array into a DataFrame
genre_encoded_df = pd.DataFrame(genre_encoded, columns=mlb.classes_)

# Step 4: Concatenate the encoded features with the original DataFrame if needed
data1 = pd.concat([data1, genre_encoded_df], axis=1)

# Check the results
print(data1.head())
```

```python
data1.head()
```

```python
#Removing Columns
features = new_data.drop(columns=['anime_id', 'name', 'type', 'episodes','genre'])
```

```python
data1.head()
```

```python
#Applying Normalization on columns Rating and Members
scaler = MinMaxScaler()
data1[['rating', 'members']] = scaler.fit_transform(data1[['rating', 'members']])
```

```python
data1.head()
```

```python
# Using LabelEncoding on categorical Features
data1 = data1.drop(['genre','genre_list','episodes'],axis = 1)
lb = LabelEncoder()
data1['type'] = lb.fit_transform(data1['type'])
for column in data1.columns:
    if data1[column].dtype == 'object':  # Categorical column
        data1[column] = data1[column].fillna(data1[column].mode()[0])
    else:  # Numeric column
        data1[column] = data1[column].fillna(data1[column].median())
```

```python
for column in features.columns:
    features[column] = features[column].fillna(features[column].mean())
similarity_matrix = cosine_similarity(features)#Calculate cosine_similarity matrix
```

```python
def recommend_anime(target_anime, data1, similarity_matrix, top_n=5):
    # Find the index of the target anime
    target_index = data1[data1['name'] == target_anime].index[0]
    
    # Get similarity scores for the target anime
    similarity_scores = list(enumerate(similarity_matrix[target_index]))
    
    # Sort by similarity score in descending order
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    
    # Exclude the target anime itself and pick top_n recommendations
    recommendations = [data1.iloc[i[0]]['name'] for i in similarity_scores[1:top_n+1]]
    
    return recommendations
```

```python
train, test = train_test_split(data1, test_size=0.2, random_state=42)
```

```python
def evaluate_recommendation_system(train, test, similarity_matrix):
    precision_list, recall_list = [], []
    
    for _, row in test.iterrows():
        recommendations = recommend_anime(row['name'], train, similarity_matrix)
        relevant_items = set(train[train['name'] == row['name']]['genres'][0])
        recommended_items = set(recommendations)
        
        # Calculate precision and recall
        true_positives = len(relevant_items & recommended_items)
        precision = true_positives / len(recommended_items) if recommended_items else 0
        recall = true_positives / len(relevant_items) if relevant_items else 0
        
        precision_list.append(precision)
        recall_list.append(recall)
    
    # Average metrics
    avg_precision = sum(precision_list) / len(precision_list)
    avg_recall = sum(recall_list) / len(recall_list)
    f1_score = 2 * (avg_precision * avg_recall) / (avg_precision + avg_recall) if (avg_precision + avg_recall) else 0
    
    return avg_precision, avg_recall, f1_score
```

```python

recommendations = recommend_anime('Naruto', train, similarity_matrix)
print("Recommendations for Naruto:", recommendations)
```
